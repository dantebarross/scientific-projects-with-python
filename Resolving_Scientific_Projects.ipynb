# Coding a Sierpinski triangle!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

# inicializando vetores x e y
sx = np.zeros(N)
sy = np.zeros(N)
sx
```

```python
for i in range(1, N):
  # gerando números randômicos no conjunto {1, 2, 3}
  k = np.random.randint(1, 4)
  sx[i] = sx[i - 1] / 2 + k - 1
  sy[i] = sy[i - 1] / 2

  if k == 2:
    sy[i] += 2

plt.plot(sx, sy, 'k.', markersize=1)
plt.show()
```

```python
# Pythagorean triple: special combination of three numbers, three sides that satisfies the Pythagorean theorem.
# A triangle whose sides form a Pythagorean triple is called a Pythagorean triangle, and is necessarily a right triangle.

def check_pythagorean_triple(a, b, c):
  if c**2 == a**2 + b**2:
    print("It's a pythagorean triple!")
  else:
    print("It's not a pythagorean triple :(")
```

```python
# Exercise: creating function that computes factorial of integers
# 5! = 5*4*3*2*1

def get_factorial(integer):
  array = np.array(range(1, integer))
  factorial = integer

  for number in array[::-1]:
    factorial = factorial * number

  return factorial
```

```python
%%time

get_factorial(5)
```

```python
# OOP (classes)

class Experiment(object): # this is the super class
  # constructor method -> is how the object is created
  def __init__(self, eType, dataN, status='ongoing'):
          # when creating methods inside of a class, first input is always self, cause when we are starting to build up this class
          # and this particular variable (object), we need to input this particular variable (object) into the function
    self.expType = eType # now we are assigning a particular piece of information to a fild (which is like a key in a dictionary)
    self.dataCount = dataN # we are using different names, but we can make these be the same:
    self.status = status
  
  def increaseN(self, number):
    self.dataCount += number
  
  def changeExpStatus(self, new_status):
    self.status = new_status

  def __str__(self):
    return 'Experiment type: ' + self.expType + ' has ' + str(self.dataCount) + ' data points.'
```

```python
## creating dot product function: element-wise multiplication of two vectors + some
## [1, 2], [3, 0] -> 1*3 + 2*0
# function checks that both inputs are numpy array (hint: isinstance())
# function checks for same lenght of two input vectors
#returns dot product if pass, useful error message otherwise (hint: raise)
```

```python
def dot_product(first_vector, second_vector):
  if not (isinstance(first_vector, np.ndarray) & isinstance(second_vector, np.ndarray)):
    raise Exception("One of the parameters isn't a numpy array")
    
  if not len(first_vector) == len(second_vector):
    raise Exception("The numpy arrays isn't the same lenght")
      
  result = 0
  for index, _ in enumerate(first_vector):
    result = result + (first_vector[index] * second_vector[index])
  
  return result

def dot_product_resolution(first_vector, second_vector):
  if not (isinstance(first_vector, np.ndarray) & isinstance(second_vector, np.ndarray)):
    raise Exception("One of the parameters isn't a numpy array")
    
  if not len(first_vector) == len(second_vector):
    raise Exception("The numpy arrays isn't the same lenght")
      
  return sum(first_vector * second_vector)
```

```python
%%time

dot_product(np.array([1, 2, 3]), np.array([4, 5, 6]))
```

```python
%%time

dot_product_resolution(np.array([1, 2, 3]), np.array([4, 5, 6]))
```

# Fibonacci series for n+3 numbers

```python
# Let's create a function that reports the Fibonacci series for n+3 numbers
## 0, 1, 1 -> 1 + 1 is 2, so:
## 0, 1, 1, 2, then:
### 0, 1, 1, 2, 3, 5

def fibonacci_series(n):
  fibonacci_serie = [0, 1, 1]
  for i in range(n):
    fibonacci_serie.append(fibonacci_serie[-2] + fibonacci_serie[-1])
  return fibonacci_serie
```

```python
fibonacci_series(10)
```

```python
# Exercise: make a Poisson-like counter.
# parameter lambda -> positive number
# so, specify some positive number that is lambda, and here you can use number 10 for example
# start at 1, multiply 1 by random number uniform([0,1])
# so, everytime you pick a number at random between zero and one, and multiply it by one, then its going to 
# get a bit smaller. Example: 1* 0.9 = 0.9 -> 0.9 * 0.45 = 0.405
# target number is smaller than exp(-lambda)
# what is the number of iterations to get this target number? ( target < exp(-lambda) )

def poisson_like_counter(lbda):

  if not lbda < 0:
    actual_value = 1 # starts multiplying by one
    counter = 0
    target = np.exp(-lbda)

    while actual_value > target:
      actual_value = actual_value * np.random.uniform(0, 1)
      counter += 1

    return actual_value, counter
  
  else:
    raise Exception("Parameter isn't a positive number")
```

```python
poisson_like_counter(10)
```
